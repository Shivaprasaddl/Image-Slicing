# Scale by Scale Cropping the images

### Alphabatic Extraction from images

```python
import os
import cv2
pt = './Alpha/'
import string

# Alphabatics
uppercase_alphabet = string.ascii_uppercase
cls = uppercase_alphabet

for i in os.listdir(pt):
    im = cv2.imread(pt+i)
    a,b,_ = im.shape
    w,h = round(b/26),round(a/22) # width and height of each Cell
    for j,S in zip(range(0,b,round(b/26)),cls):
        sp = './Cropdata/'+S
        os.makedirs(sp,exist_ok=True)
        s2 = 0
        for k in range(0,a,round(a/22)):
            c = im[k:k+h,j:j+w]
            cv2.imwrite(sp+'/{}_{}.jpg'.format(i.split('.')[0],s2),c)
            s2 += 1
    print(i,' Completed...')
```

### Numbers Extraction from images 

```python
import os
import cv2
pt = './Number/'
import string

# Numbers
cls = list(range(10))

for i in os.listdir(pt):
    im = cv2.imread(pt+i)
    a,b,_ = im.shape
    w,h = round(b/10),round(a/22)
    for j,S in zip(range(0,b,round(b/10)),cls):
        sp = './Cropdata/'+str(S)
        os.makedirs(sp,exist_ok=True)
        s2 = 0
        for k in range(0,a,round(a/22)):
            c = im[k:k+h,j:j+w]
            cv2.imwrite(sp+'/{}_{}.jpg'.format(i.split('.')[0],s2),c)
            s2 += 1
    print(i,' Completed...')
```
